**Зададание 1:**  Заполните методы класса Line, чтобы они принимали координаты в виде пары 
кортежей и возвращали наклон и расстояние линии.  
class Line:  
    def \__init__\(self,coor1,coor2):  
    pass  
    def distance(self):  
    pass  
    def slope(self):  
    Pass  

```python
import math

class Line:
    def __init__(self, coor1, coor2):
        self.x1, self.y1 = coor1
        self.x2, self.y2 = coor2
    
    def distance(self):
        return math.sqrt((self.x2 - self.x1)**2 + (self.y2 - self.y1)**2)
    
    def slope(self):
        if self.x2 - self.x1 == 0:
            return None  # Вертикальная линия
        return (self.y2 - self.y1) / (self.x2 - self.x1)

line = Line((3, 2), (8, 10))
print(f"Расстояние: {line.distance():.2f}")
print(f"Наклон: {line.slope():.2f}")
```

**Зададание 2:** Заполните класс Cylinder:  
class Cylinder:  
def \__init__\(self,height=1,radius=1):  
pass  
def volume(self):  
pass  
def surface_area(self):  
pass 

```python
class Cylinder:
    def __init__(self, height=1, radius=1):
        self.height = height
        self.radius = radius
    
    def volume(self):
        return math.pi * self.radius**2 * self.height
    
    def surface_area(self):
        return 2 * math.pi * self.radius * (self.radius + self.height)

cyl = Cylinder(2, 3)
print(f"Объем: {cyl.volume():.2f}")
print(f"Площадь поверхности: {cyl.surface_area():.2f}")
```

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
    
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Внесено {amount}. Новый баланс: {self.balance}")
        return self.balance
    
    def withdraw(self, amount):
        if amount > self.balance:
            print(f"Недостаточно средств! Баланс: {self.balance}")
            return self.balance
        self.balance -= amount
        print(f"Снято {amount}. Новый баланс: {self.balance}")
        return self.balance
    
    def __str__(self):
        return f"Владелец: {self.owner}, Баланс: {self.balance}"

test = BankAccount("Иван Иванов", 1000)
print(test)
test.deposit(500)
test.withdraw(200)
test.withdraw(2000)  # something wrong
```

**Зададание 4:** Product Inventory Project – создайте приложение, которое управляет 
инвентаризацией продуктов. Создайте класс продукта, у которого есть цена, 
идентификатор и количество. Затем создайте класс запасов, который отслеживает 
различные продукты и может суммировать стоимость запасов. 

```python
class Product:
    def __init__(self, id, name, price, quantity):
        self.id = id
        self.name = name
        self.price = price
        self.quantity = quantity
    
    def get_value(self):
        return self.price * self.quantity
    
    def __str__(self):
        return f"{self.name} (ID: {self.id}) - {self.price} руб. x {self.quantity} = {self.get_value()} руб."

class Inventory:
    def __init__(self):
        self.products = {}
    
    def add_product(self, product):
        self.products[product.id] = product
    
    def remove_product(self, product_id):
        if product_id in self.products:
            del self.products[product_id]
    
    def total_value(self):
        return sum(product.get_value() for product in self.products.values())
    
    def __str__(self):
        result = "Затарчик:\n"
        for product in self.products.values():
            result += f"  {product}\n"
        result += f"Общая стоимость: {self.total_value()} руб."
        return result

inv = Inventory()
inv.add_product(Product(1, "Яблоки", 50, 100))
inv.add_product(Product(2, "Бананы", 70, 50))
inv.add_product(Product(3, "Орэшки(цена за штучку!)", 10, 60))
print(inv)
print("Бананы не понравились!")
inv.remove_product(2)
print(inv)
print("Орэшки понравились очень сильно! Возьму ещё :)")
inv.add_product(Product(3, "Орэшки(цена за штучку!)", 10, 100))
print(inv)
```

**Зададание 5:** Система бронирования авиакомпаний/отелей. Создайте систему бронирования, 
которая бронирует места в самолетах или номера в отелях. Он взимает различные тарифы 
за определенные участки самолета или отеля. Например, первый класс будет стоить 
больше, чем эконом. В гостиничных номерах есть пентхаусы, которые стоят дороже. 
Следите за тем, когда комнаты будут доступны и могут быть запланированы.  

```python
class Room:
    def __init__(self, number, room_type, price):
        self.number = number
        self.type = room_type
        self.price = price
        self.is_booked = False
    
    def book(self):
        if not self.is_booked:
            self.is_booked = True
            return True
        return False
    
    def release(self):
        self.is_booked = False
    
    def __str__(self):
        status = "забронирован" if self.is_booked else "свободен"
        return f"Комната {self.number} ({self.type}) - {self.price} руб. - {status}"

class Hotel:
    def __init__(self, name):
        self.name = name
        self.rooms = {}
    
    def add_room(self, room):
        self.rooms[room.number] = room
    
    def book_room(self, room_number):
        if room_number in self.rooms:
            return self.rooms[room_number].book()
        return False
    
    def available_rooms(self):
        return [room for room in self.rooms.values() if not room.is_booked]
    
    def __str__(self):
        result = f"Отель: {self.name}\n"
        for room in self.rooms.values():
            result += f"  {room}\n"
        return result

hotel = Hotel("Отель Элеон(10 сезон)")
hotel.add_room(Room(101, "Стандарт", 2000))
hotel.add_room(Room(102, "Люкс", 5000))
hotel.add_room(Room(201, "Пентхаус", 10000))

print(hotel)
hotel.book_room(101)
print("\nПосле бронирования комнаты 101:")
print(hotel)
```

**Зададание 6:** Менеджер компании – создайте иерархию классов – абстрактный класс Employee и 
подклассы HourlyEmployee, SalariedEmployee, Manager и Executive. Зарплата у всех 
рассчитывается по-разному, немного об этом узнайте. После того, как вы установили 
иерархию сотрудников, создайте класс Company, который позволит вам управлять 
сотрудниками. Вы должны иметь возможность нанимать, увольнять и повышать 
сотрудников.  

```python
from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, name, id):
        self.name = name
        self.id = id
    
    @abstractmethod
    def calculate_salary(self):
        pass
    
    def __str__(self):
        return f"{self.name} (ID: {self.id}) - Зарплата: {self.calculate_salary()} руб."

class HourlyEmployee(Employee):
    def __init__(self, name, id, hourly_rate, hours_worked):
        super().__init__(name, id)
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked
    
    def calculate_salary(self):
        return self.hourly_rate * self.hours_worked

class SalariedEmployee(Employee):
    def __init__(self, name, id, monthly_salary):
        super().__init__(name, id)
        self.monthly_salary = monthly_salary
    
    def calculate_salary(self):
        return self.monthly_salary

class Manager(SalariedEmployee):
    def __init__(self, name, id, monthly_salary, bonus):
        super().__init__(name, id, monthly_salary)
        self.bonus = bonus
    
    def calculate_salary(self):
        return self.monthly_salary + self.bonus

class Executive(Manager):
    def __init__(self, name, id, monthly_salary, bonus, stock_options):
        super().__init__(name, id, monthly_salary, bonus)
        self.stock_options = stock_options
    
    def calculate_salary(self):
        return super().calculate_salary() + self.stock_options

class Company:
    def __init__(self, name):
        self.name = name
        self.employees = {}
        self.next_id = 1
    
    def hire(self, employee):
        employee.id = self.next_id
        self.employees[self.next_id] = employee
        self.next_id += 1
        print(f"Нанят: {employee.name}")
    
    def fire(self, employee_id):
        if employee_id in self.employees:
            employee = self.employees[employee_id]
            del self.employees[employee_id]
            print(f"Уволен: {employee.name}")
    
    def promote(self, employee_id, new_position):
        if employee_id in self.employees:
            print(f"Повышен: {self.employees[employee_id].name} -> {new_position}")
    
    def total_payroll(self):
        return sum(emp.calculate_salary() for emp in self.employees.values())
    
    def __str__(self):
        result = f"Компания: {self.name}\n"
        for emp in self.employees.values():
            result += f"  {emp}\n"
        result += f"Общий фонд зарплат: {self.total_payroll()} руб."
        return result

company = Company("Элтекс")
company.hire(HourlyEmployee("Bob", 0, 500, 160))
company.hire(SalariedEmployee("Ejji", 0, 80000))
company.hire(Manager("Silva", 0, 120000, 20000))
print(company)

company.fire(3) # Сильва выстрелил себе в ногу:(
print(company)

company.promote(1, 'senior: работы больше, зарплата та же :0')
print(company)
```

**Зададание 7:** Планировщик пациентов/врачей. Создайте класс пациента и класс врача. Наймите 
врача, который может обслуживать нескольких пациентов, и настройте программу 
планирования, согласно которой врач может обслуживать только 16 пациентов в течение 
8-часового рабочего дня.  

```python
class Patient:
    def __init__(self, name, medical_history=""):
        self.name = name
        self.medical_history = medical_history
    
    def __str__(self):
        return f"Пациент: {self.name}"

class Doctor:
    def __init__(self, name, specialty):
        self.name = name
        self.specialty = specialty
        self.schedule = {}  # {время: пациент}
        self.work_hours = (9, 17)
        self.appointment_duration = 30  # минут
    
    def available_slots(self):
        start_hour, end_hour = self.work_hours
        slots = []
        current_time = start_hour * 60  # в минутах
        end_time = end_hour * 60
        
        while current_time + self.appointment_duration <= end_time:
            time_str = f"{current_time // 60:02d}:{current_time % 60:02d}"
            if time_str not in self.schedule:
                slots.append(time_str)
            current_time += self.appointment_duration
        return slots
    
    def book_appointment(self, patient, time_slot):
        if time_slot in self.available_slots():
            self.schedule[time_slot] = patient
            print(f"Забронировано: {patient.name} на {time_slot}")
            return True
        print(f"Время {time_slot} недоступно")
        return False
    
    def __str__(self):
        result = f"Доктор: {self.name} ({self.specialty})\n"
        result += "Расписание:\n"
        for time, patient in sorted(self.schedule.items()):
            result += f"  {time}: {patient.name}\n"
        return result

doc = Doctor("Lola", "Психотерапевт") # ребята из прошлой задачи пришли лечиться, ведь они пишут на C++
p1 = Patient("Ejji")
p2 = Patient("Bob")

doc.book_appointment(p1, "10:00")
doc.book_appointment(p2, "10:30")
print(doc)
print("Свободные слоты:", doc.available_slots()[:3]) # 7 - кол-во свободных слотов на экране, можно больше, но не надо
```

**Зададание 8:** Создатель и менеджер рецептов. Создайте класс рецептов с ингредиентами и 
поместите их в программу менеджера рецептов, которая организует их по категориям, 
таким как десерты, основные блюда, или по ингредиентам, таким как курица, говядина, 
супы, пироги и т. д.  

```python
class Recipe:
    def __init__(self, name, category, ingredients, instructions):
        self.name = name
        self.category = category
        self.ingredients = ingredients
        self.instructions = instructions
    
    def contains_ingredient(self, ingredient):
        return any(ingredient.lower() in ing.lower() for ing in self.ingredients)
    
    def __str__(self):
        return f"{self.name} ({self.category})\nИнгредиенты: {', '.join(self.ingredients)}\nРецептик: {(self.instructions)}"

class RecipeManager:
    def __init__(self):
        self.recipes = []
    
    def add_recipe(self, recipe):
        self.recipes.append(recipe)
    
    def get_by_category(self, category):
        return [recipe for recipe in self.recipes if recipe.category.lower() == category.lower()]
    
    def get_by_ingredient(self, ingredient):
        return [recipe for recipe in self.recipes if recipe.contains_ingredient(ingredient)]
    
    def __str__(self):
        result = "Рецепты:\n"
        for recipe in self.recipes:
            result += f"  {recipe}\n"
        return result

manager = RecipeManager()
manager.add_recipe(Recipe("Омлет", "Завтрак", ["яйца", "молоко", "соль"], "Взбить и пожарить"))
manager.add_recipe(Recipe("Суп", "Обед", ["курица", "овощи", "вода"], "Варить 30 минут"))
manager.add_recipe(Recipe("Студент живущий в общаге", "Весь день", ["гречка"], "Варить 20 минут (по желанию)"))

print("Все рецепты:")
print(manager)
print("Завтраки:", [r.name for r in manager.get_by_category("завтрак")])
print("С курицей:", [r.name for r in manager.get_by_ingredient("курица")])
```

**Зададание 9:** Создатель семейного древа. Создайте класс под названием «Человек», у которого 
будет имя, когда он родился и когда (и если) умер. Разрешить пользователю создавать эти 
классы Person и помещать их в структуру генеалогического дерева. Распечатайте дерево на 
экран.  

```python
from datetime import datetime, date

class Person:
    def __init__(self, name, birth_date, death_date=None):
        self.name = name
        self.birth_date = birth_date
        self.death_date = death_date
        self.children = []
        self.parent = None
    
    def add_child(self, child):
        child.parent = self
        self.children.append(child)
    
    def is_alive(self):
        return self.death_date is None
    
    def get_age(self):
        end_date = self.death_date or date.today()
        return end_date.year - self.birth_date.year - ((end_date.month, end_date.day) < (self.birth_date.month, self.birth_date.day))
    
    def __str__(self):
        status = "жив" if self.is_alive() else f"умер в {self.death_date.year}"
        return f"{self.name} ({self.birth_date.year}-{status})"

def print_family_tree(person, level=0):
    indent = "    " * level
    print(f"{indent}{person}")
    for child in person.children:
        print_family_tree(child, level + 1)

grandpa = Person("Иван", date(1950, 5, 15))
father = Person("Петр", date(1975, 3, 10))
mother = Person("Мария", date(1978, 8, 20))
child1 = Person("Анна", date(2000, 12, 5))
child2 = Person("Алексей", date(2005, 7, 15))

grandpa.add_child(father)
father.add_child(child1)
father.add_child(child2)

print("Семейное древо:")
print_family_tree(grandpa)
```

**Зададание 10:**  Создайте класс игральных карт, а далее класс колоды с этими картами. В колоде 52 
карты, каждая из которых принадлежит одной из четырех мастей и имееет одно из 13 
значений. Масти такие: пики (англ.: Spades), червы (англ.: Hearts), бубны (англ.: Diamonds) 
и трефы (англ.: Clubs). Значения такие: туз (англ.: Ace), 2, 3, 4, 5, 6, 7, 8, 9, 10, валет (англ.: 
Jack), дама (англ.: Queen) и король (англ.: King). Должна быть возможность тасовать колоду, 
извлекать и сдавать карты.  

```python
from random import shuffle

class Card:
    suits = ["Пики", "Червы", "Бубны", "Трефы"]
    values = ["2", "3", "4", "5", "6", "7", "8", "9", "10", "Валет", "Дама", "Король", "Туз"]
    
    def __init__(self, suit, value):
        self.suit = suit
        self.value = value
    
    def __str__(self):
        return f"{self.value} {self.suit}"

class Deck:
    def __init__(self):
        self.cards = []
        self.create_deck()
    
    def create_deck(self):
        for suit in Card.suits:
            for value in Card.values:
                self.cards.append(Card(suit, value))
    
    def shuffle(self):
        shuffle(self.cards)
    
    def deal(self, num_cards=1):
        if len(self.cards) < num_cards:
            return None
        return [self.cards.pop() for _ in range(num_cards)]
    
    def __str__(self):
        return f"Колода: {len(self.cards)} карт"

deck = Deck()
print(deck)
deck.shuffle()
hand = deck.deal(5)
print("Раздано 5 карт:", [str(card) for card in hand])
print(f"Осталось в колоде: {len(deck.cards)} карт")
deck.shuffle()
hand = deck.deal(7)
print("Раздано 7 карт:", [str(card) for card in hand])
```
